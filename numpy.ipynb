```python
import numpy as np

# Numpy - Fast Number Crunching
a = np.arange(1_000_000, dtype=np.float64)
b = np.sqrt(a) * 2.5 + np.log1p(a)

# reduce to a single scalar
result = np.sum(b)
result
```
