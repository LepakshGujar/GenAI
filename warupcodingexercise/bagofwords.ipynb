```python
from collections import Counter
import re

def text_to_word_frequency(text):
    """
    Convert text to a frequency count of words.
    Ignores order, just counts word occurrence.
    
    Args:
        text (str): Input text
        
    Returns:
        dict: Dictionary with words as keys and frequencies as values
    """
    # Convert to lowercase and remove punctuation
    text = text.lower()
    words = re.findall(r'\b\w+\b', text)
    
    # Count word frequencies
    word_freq = Counter(words)
    
    return dict(word_freq)

# Example usage
sample_text = "Hello world, hello Python! World of programming."
result = text_to_word_frequency(sample_text)
print(result)
```
