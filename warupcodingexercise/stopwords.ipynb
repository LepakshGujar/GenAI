```python
# Define common stopwords
stopwords = {'the', 'is', 'a', 'an', 'in', 'and', 'or', 'but', 'to', 'of', 'for', 'on', 'it', 'at', 'by', 'from', 'with', 'as', 'be', 'are', 'was', 'were', 'been', 'being'}

# Function to remove stopwords from text
def remove_stopwords(text):
    """Remove common stopwords from text."""
    words = text.lower().split()
    filtered_words = [word for word in words if word not in stopwords]
    return ' '.join(filtered_words)

# Example usage
sample_text = "The quick brown fox is in the lazy garden"
result = remove_stopwords(sample_text)
print(f"Original: {sample_text}")
print(f"Filtered: {result}")
```
