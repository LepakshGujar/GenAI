```python
import re
from collections import Counter

class TextAnalyzer:
    def __init__(self, text):
        self.text = text
        self.words = self._extract_words()
    
    def _extract_words(self):
        """Extract words from text (lowercase, alphanumeric only)"""
        words = re.findall(r'\b[a-z]+\b', self.text.lower())
        return words
    
    def word_count(self):
        """Return total number of words"""
        return len(self.words)
    
    def unique_words(self):
        """Return number of unique words"""
        return len(set(self.words))
    
    def char_count(self):
        """Return total character count (excluding spaces)"""
        return len(self.text.replace(" ", ""))
    
    def word_frequency(self, top_n=10):
        """Return most common words"""
        return Counter(self.words).most_common(top_n)
    
    def avg_word_length(self):
        """Return average word length"""
        return round(sum(len(w) for w in self.words) / len(self.words), 2) if self.words else 0
    
    def summary(self):
        """Print analysis summary"""
        print(f"Total Words: {self.word_count()}")
        print(f"Unique Words: {self.unique_words()}")
        print(f"Characters: {self.char_count()}")
        print(f"Avg Word Length: {self.avg_word_length()}")
        print(f"Top 10 Words: {self.word_frequency()}")

# Example usage
text = "The quick brown fox jumps over the lazy dog. The dog was really lazy."
analyzer = TextAnalyzer(text)
analyzer.summary()
```

```python
def tokenize_and_clean(sentence):
    """Tokenize a sentence into words and remove punctuation"""
    words = re.findall(r'\b[a-z]+\b', sentence.lower())
    return words

# Example usage
sample_sentence = "Hello, World! How are you?"
tokens = tokenize_and_clean(sample_sentence)
print(tokens)
```

```python
def jaccard_similarity(sentence1, sentence2):
    """Calculate Jaccard similarity between two sentences"""
    words1 = set(tokenize_and_clean(sentence1))
    words2 = set(tokenize_and_clean(sentence2))
    
    if not words1 and not words2:
        return 1.0
    
    intersection = words1.intersection(words2)
    union = words1.union(words2)
    
    similarity = len(intersection) / len(union)
    return round(similarity, 2)

# Example usage
sentence_a = "The quick brown fox jumps over the lazy dog"
sentence_b = "A fast brown fox leaps over a lazy dog"
similarity_score = jaccard_similarity(sentence_a, sentence_b)
print(f"Jaccard Similarity: {similarity_score}")
```

```python
def classify_sentence(sentence):
    """Classify a sentence as a question, exclamation, or statement"""
    sentence = sentence.strip()
    
    if sentence.endswith('?'):
        return 'question'
    elif sentence.endswith('!'):
        return 'exclamation'
    else:
        return 'statement'

# Example usage
test_sentences = [
    "How are you?",
    "This is amazing!",
    "The weather is nice today",
    sample_sentence,
    sentence_a
]

for sent in test_sentences:
    label = classify_sentence(sent)
    print(f"'{sent}' -> {label}")
```
