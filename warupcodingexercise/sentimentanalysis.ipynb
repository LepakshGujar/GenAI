```python
from transformers import pipeline

# Initialize the sentiment analysis pipeline
sentiment_analyzer = pipeline("sentiment-analysis")

def analyze_sentiment(text):
    """
    Determines whether text expresses positive, negative, or neutral sentiment.
    
    Args:
        text (str): The text to analyze
        
    Returns:
        dict: Contains 'label' (POSITIVE, NEGATIVE, NEUTRAL) and 'score' (confidence)
    """
    result = sentiment_analyzer(text)[0]
    return {
        'sentiment': result['label'],
        'confidence': result['score']
    }

# Example usage
sample_texts = [
    "I love this product! It's amazing.",
    "This is terrible and I hate it.",
    "The weather is cloudy today."
]

for text in sample_texts:
    result = analyze_sentiment(text)
    print(f"Text: {text}")
    print(f"Sentiment: {result['sentiment']}, Confidence: {result['confidence']:.4f}\n")
```
