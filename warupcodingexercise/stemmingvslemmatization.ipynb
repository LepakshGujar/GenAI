```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
import nltk

# Download required NLTK data
nltk.download('wordnet')
nltk.download('omw-1.4')

# Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Example words
words = ['running', 'runs', 'ran', 'easily', 'fairly', 'studies', 'studying']

print("Word\t\t| Stemming\t| Lemmatization")
print("-" * 50)

for word in words:
    stem = stemmer.stem(word)
    lemma = lemmatizer.lemmatize(word, pos='v')  # 'v' for verb
    print(f"{word}\t\t| {stem}\t\t| {lemma}")

print("\n\nKey Differences:")
print("- Stemming: Removes suffixes using rules (faster, may not be valid words)")
print("- Lemmatization: Returns dictionary-based root form (slower, always valid words)")
```
