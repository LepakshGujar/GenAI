```python
# generate data types and variables

integer_var = 42
float_var = 3.14
string_var = "hello"
boolean_var = True
list_var = [1, 2, 3]
tuple_var = (4, 5, 6)
dict_var = {"name": "Alice", "age": 30}
set_var = {"apple", "banana", "cherry"}
none_var = None

print(integer_var, float_var, string_var, boolean_var)
print(list_var, tuple_var, dict_var, set_var, none_var)
```

```python
if boolean_var:
    print("boolean_var is True")

if integer_var > float_var:
    print(f"integer_var ({integer_var}) is greater than float_var ({float_var})")

if dict_var.get("age") == 30:
    print(f"{dict_var['name']} is {dict_var['age']} years old")

if list_var:
    print("list_var has", len(list_var), "elements:", list_var)

if none_var is None:
    print("none_var is None")

if set_var:
    print("set_var contains", set_var)

if tuple_var[0] == 4:
    print("tuple_var starts with 4")
```

```python
if boolean_var:
    print("boolean_var is True")
else:
    print("boolean_var is False")

if integer_var > float_var:
    print(f"{integer_var} is greater than {float_var}")
else:
    print(f"{integer_var} is not greater than {float_var}")
    
```

```python
if integer_var > 50:
    print("integer_var is greater than 50")
elif integer_var > 10:
    print("integer_var is greater than 10 but not greater than 50")
else:
    print("integer_var is 10 or less")

if boolean_var:
    print("boolean_var is True")
elif none_var is None:
    print("none_var is None")
else:
    print("boolean_var is False and none_var is not None")

if len(list_var) == 0:
    print("list_var is empty")
elif len(list_var) < 5:
    print("list_var has fewer than 5 elements")
else:
    print("list_var has 5 or more elements")
```

```python
count = 0
while count < len(list_var):
    print(f"list_var[{count}] = {list_var[count]}")
    count += 1
```

```python
for i, value in enumerate(list_var):
    print(f"list_var[{i}] = {value}")
```

```python
count = 0
while True:
    print(f"list_var[{count}] = {list_var[count]}")
    count += 1
    if count >= len(list_var):
        break
```
